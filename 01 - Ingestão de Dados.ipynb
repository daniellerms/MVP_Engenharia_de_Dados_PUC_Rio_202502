```python
#O arquivo csv "MICRODADOS_ENEM_2023.csv" está disponível no site do INEP url:  https://www.gov.br/inep/pt-br/acesso-a-informacao/dados-abertos/microdados/enem

# arquivo foi salvo no volume default do databricks conforme indicado na variável caminho_arquivo

caminho_arquivo = "/Volumes/workspace/default/mvp/MICRODADOS_ENEM_2023.csv"

#leitura e criação do dataframe do csv usando spark
df_microdadosenem2023 = spark.read.csv(caminho_arquivo, header=True, inferSchema=True,sep=";",encoding = "ISO-8859-1")

#visualização dos dados
display(df_microdadosenem2023)



```

```python
#nome das colunas do dataframe
lista_de_colunas = df_microdadosenem2023.columns

print(lista_de_colunas)
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS bronze;
```

```python
#criando a versão bronze da tabela no schema bronze
df_microdadosenem2023.write.format("delta").mode("overwrite").saveAsTable("bronze.bronze_enem2023")
```
